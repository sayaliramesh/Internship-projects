## Task 7: Common Information Sources
### Objective: Analyze common sources participants rely on for investment information.

```python
df['Source'].value_counts()
```

### Interpretation

The analysis shows that most participants rely on financial consultants (16) for investment information, making it the most common source. ## Task 8: Investment Duration
### Objective: Calculate the average investment duration.

```python
df['Duration'].value_counts()
```

```python
duration_map = {
    'Less than 1 year': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    'More than 5 years': 6
}

df['Duration_numeric'] = df['Duration'].map(duration_map)

average_duration = df['Duration_numeric'].mean()

print("Average Investment Duration:", average_duration)
```

### Interpretation

The calculated average investment duration is 2.98 years. This indicates that most participants prefer medium-term investments, typically around 3 years. It suggests that investors aim to balance risk and returns while maintaining a moderate investment horizon.

## Task 9: Expectations from Investments
### Objective: Summarize participants' expectations from investments

```python
df['Expect'].value_counts()
```

### Interpretation

The analysis of the Expect column shows that most participants expect returns between 20%–30% (32 participants), making it the most common expectation. A smaller group expects 30%–40% returns (5 participants), while 3 participants expect 10%–20% returns. 
Overall, the results indicate that the majority of investors have moderately high return expectations from their investments.

## Task 10: Correlation Analysis
### Objective: Explore potential correlations between factors.

```python
# Convert Duration to numeric
duration_map = {
    'Less than 1 year': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    'More than 5 years': 6}

df['Duration_num'] = df['Duration'].map(duration_map)

# Convert Expected Returns to numeric
expect_map = {
    '10%-20%': 15,
    '20%-30%': 25,
    '30%-40%': 35}

df['Expect_num'] = df['Expect'].map(expect_map)
```

```python
df[['age','Duration_num','Expect_num']].corr()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['age'], df['Expect_num'])
plt.xlabel("Age")
plt.ylabel("Expected Returns (%)")
plt.title("Age vs Expected Returns")
plt.show()
```
