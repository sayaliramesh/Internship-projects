```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

## Level 1 - Task 1: Top 3 Most Common Cuisines

In this task, we identify the top three cuisines served in the dataset and calculate their percentage contribution.

```python
# Split cuisines 
df['Cuisines'] = df['Cuisines'].astype(str)

cuisine_series = df['Cuisines'].str.split(', ').explode()

top_cuisines = cuisine_series.value_counts().head(3)

# Percentage calculation
top_cuisine_percentage = (top_cuisines / len(df)) * 100

print("Top 3 Cuisines:\n")
print(top_cuisines)

print("\nPercentage of Top 3 Cuisines:\n")
print(top_cuisine_percentage)
```

```python
plt.figure()
top_cuisines.plot(kind='bar')
plt.title("Top 3 Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Number of Restaurants")
plt.show()
```

### Insight:

North Indian cuisine dominates the market with 41.46% share, followed by Chinese (28.63%) and Fast Food (20.79%). 


## Level 1 - Task 2: City-wise Restaurant Analysis

This task identifies:
- The city with the highest number of restaurants.
- The average rating per city.
- The city with the highest average rating.

```python
# City with highest number of restaurants
top_city = df['City'].value_counts().idxmax()
print("City with Highest Restaurants:", top_city)

# Average rating per city
city_avg_rating = df.groupby('City')['Aggregate rating'].mean()

# City with highest average rating
highest_rated_city = city_avg_rating.idxmax()
print("City with Highest Average Rating:", highest_rated_city)
```

```python
plt.figure()
city_avg_rating.sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top 10 Cities by Average Rating")
plt.ylabel("Average Rating")
plt.show()
```

### Insight:

New Delhi has the highest number of restaurants, indicating it is the most saturated and competitive market in the dataset.  
However, Inner City records the highest average rating, suggesting stronger customer satisfaction and potentially better service quality in that region.

## Level 1 - Task 3: Price Range Distribution

This task analyzes how restaurants are distributed across different price categories.

```python
price_distribution = df['Price range'].value_counts()

price_percentage = (price_distribution / len(df)) * 100

print("Price Distribution:\n", price_distribution)
print("\nPrice Percentage:\n", price_percentage)
```

```python
plt.figure()
price_distribution.plot(kind='bar')
plt.title("Price Range Distribution")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.show()
```

### Insight:

Nearly 46.53% of restaurants fall under Price Range 1, indicating a strong dominance of budget-friendly dining options. Mid-range restaurants (32.59%) also hold a significant share, while premium categories (Price Range 3 and 4) represent a smaller portion of the market.  
This suggests that the restaurant industry in this dataset is largely driven by affordability and mass-market demand.

## Level 1 - Task 4: Online Delivery Impact Analysis

This task determines:
- Percentage of restaurants offering online delivery.
- Comparison of average ratings between delivery and non-delivery restaurants.

```python
# Percentage offering delivery
online_delivery_percentage = df['Has Online delivery'].value_counts(normalize=True) * 100
print("Online Delivery Percentage:\n", online_delivery_percentage)

# Compare ratings
delivery_rating = df.groupby('Has Online delivery')['Aggregate rating'].mean()
print("\nAverage Rating Comparison:\n", delivery_rating)
```

```python
plt.figure()
delivery_rating.plot(kind='bar')
plt.title("Average Rating: Online Delivery vs No Delivery")
plt.ylabel("Average Rating")
plt.show()
```

### Insight:

Approximately 74.34% of restaurants do not offer online delivery, while only 25.66% provide this service. However, restaurants with online delivery have a significantly higher average rating (3.25) compared to those without delivery (2.47).  
This suggests that offering online delivery may positively influence customer satisfaction and overall ratings.

## Level 2 - Task 1: Restaurant Ratings Analysis

This task analyzes:
- Distribution of aggregate ratings
- Most common rating range
- Average number of votes received

```python
# Distribution of Aggregate Ratings
plt.figure()
df['Aggregate rating'].hist(bins=20)
plt.title("Distribution of Aggregate Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Restaurants")
plt.show()

# Most common rating range
most_common_rating = df['Aggregate rating'].value_counts().idxmax()
print("Most Common Rating:", most_common_rating)

# Average votes
avg_votes = df['Votes'].mean()
print("Average Number of Votes:", round(avg_votes, 2))
```

### Insight:

The most common rating is 0.0, indicating that a significant number of restaurants are either unrated or newly listed without customer feedback.  
On average, restaurants receive approximately 157 votes, suggesting moderate customer engagement across the dataset.

## Level 2 - Task 2: Cuisine Combination Analysis

This task identifies:
- Most common cuisine combinations
- Whether certain combinations receive higher ratings

```python
# Most common cuisine combinations
combo_counts = df['Cuisines'].value_counts().head(10)
print("Top Cuisine Combinations:\n", combo_counts)

# Average rating per cuisine combination
combo_rating = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
print("\nTop Rated Cuisine Combinations:\n", combo_rating)
```

### Insight:

Single-cuisine restaurants, particularly North Indian (936), dominate the dataset, followed by popular combinations like North Indian & Chinese (511). However, the highest-rated cuisine combinations (average rating 4.9) are mostly premium or international multi-cuisine offerings such as Continental-Indian and American-Burger-Grill.  
This suggests that while traditional cuisines drive volume, diverse and global cuisine combinations tend to achieve higher customer satisfaction.

## Level 2 - Task 3: Geographic Analysis

This task visualizes restaurant locations using latitude and longitude
to identify geographic clusters.

```python
plt.figure()
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.3)
plt.title("Geographic Distribution of Restaurants")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

### Insight:

The geographic scatter plot reveals clear clustering of restaurants in specific longitude and latitude ranges, indicating strong urban concentration.  
Restaurants are densely located in particular regions, suggesting high-demand metropolitan areas, while other regions show sparse distribution.  
This pattern reflects market concentration in major cities and economically active zones.

## Level 2 - Task 4: Restaurant Chains Analysis

This task identifies restaurant chains and analyzes their ratings and popularity.

```python
# Identify chains (restaurants appearing multiple times)
chain_counts = df['Restaurant Name'].value_counts()
chains = chain_counts[chain_counts > 1].head(10)

print("Top Restaurant Chains:\n", chains)

# Average rating of chains
chain_ratings = df[df['Restaurant Name'].isin(chains.index)] \
                .groupby('Restaurant Name')['Aggregate rating'] \
                .mean() \
                .sort_values(ascending=False)

print("\nAverage Ratings of Top Chains:\n", chain_ratings)
```

### Insight:

Cafe Coffee Day and Domino’s Pizza have the highest outlet presence, indicating strong market expansion and brand penetration. However, Barbeque Nation records the highest average rating (4.35), suggesting superior customer satisfaction despite having fewer outlets.  
This shows that larger chain presence does not necessarily guarantee higher ratings, and quality perception varies across brands.

## Level 3 - Task 1: Restaurant Reviews Analysis

This task analyzes textual feedback indicators to understand customer sentiment patterns.  
We identify the most common positive and negative keywords and examine whether review length has any relationship with restaurant ratings.

```python
# Most common review words (from Rating text)
review_words = df['Rating text'].value_counts()
print("Most Common Review Keywords:\n")
print(review_words)
```

```python
# Review length
df['review_length'] = df['Rating text'].astype(str).apply(len)
avg_review_length = df['review_length'].mean()

print("Average Review Length:", round(avg_review_length,2))
```

```python
plt.figure()
sns.scatterplot(x=df['review_length'], y=df['Aggregate rating'])
plt.title("Review Length vs Rating")
plt.xlabel("Review Length")
plt.ylabel("Rating")
plt.show()
```

### Insight:

The most common review keyword is "Average", followed by "Not rated" and "Good", indicating that a large portion of restaurants receive moderate feedback or lack sufficient ratings.
Higher sentiment categories such as "Very Good" and "Excellent" appear less frequently, suggesting that only a smaller proportion of restaurants achieve very high customer satisfaction levels.

## Level 3 - Task 2: Votes Analysis

This task identifies restaurants with the highest and lowest number of votes and examines whether customer voting activity correlates with restaurant ratings.

```python
# Highest votes
highest_votes = df.loc[df['Votes'].idxmax(), ['Restaurant Name','Votes']]
print("Restaurant with Highest Votes:\n", highest_votes)

# Lowest votes
lowest_votes = df.loc[df['Votes'].idxmin(), ['Restaurant Name','Votes']]
print("\nRestaurant with Lowest Votes:\n", lowest_votes)
```

```python
correlation = df['Votes'].corr(df['Aggregate rating'])
print("\nCorrelation between Votes and Rating:", round(correlation,3))
```

```python
plt.figure()
sns.scatterplot(x=df['Votes'], y=df['Aggregate rating'])
plt.title("Votes vs Restaurant Rating")
plt.show()
```

### Insight:

The restaurant "Toit" has the highest number of votes (10,934), indicating strong customer engagement and popularity. In contrast, "Cantinho da Gula" has received no votes, suggesting very low visibility or customer interaction.
The correlation between votes and ratings is 0.314, showing a weak positive relationship. This suggests that restaurants with more votes tend to have slightly higher ratings, but popularity alone does not strongly determine customer satisfaction.

## Level 3 - Task 3: Price Range vs Online Delivery & Table Booking

This task explores whether restaurant pricing levels influence the availability of online delivery and table booking services.

```python
# Price range vs table booking
table_booking_price = pd.crosstab(df['Price range'], df['Has Table booking'])
print(table_booking_price)
```

```python
online_delivery_price = pd.crosstab(df['Price range'], df['Has Online delivery'])
online_delivery_price.plot(kind='bar', stacked=True)

plt.title("Price Range vs Online Delivery")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
table_booking_price.plot(kind='bar', stacked=True)
plt.title("Price Range vs Table Booking")
plt.show()
```

### Insight

Table booking availability increases with higher price ranges.
Price Range 1 restaurants almost never offer table booking, while Price Range 3 and 4 restaurants provide this service more frequently.
This indicates that higher-priced restaurants are more likely to offer reservation services and structured dining experiences.
